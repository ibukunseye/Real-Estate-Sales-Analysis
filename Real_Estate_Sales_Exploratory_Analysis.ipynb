## Data Maniupulation and Validation

```python
# fill the missing values of 'Address', 'Property Type', 'Residential type' with 'Unknown'
df['Address'] = df['Address'].astype(str).fillna('Unknown')
df['Address'] = df['Address'].replace('nan', 'Unknown')

df['Property Type'] = df['Property Type'].astype(str).fillna('Unknown')
df['Property Type'] = df['Property Type'].replace('nan', 'Unknown')

df['Residential Type'] = df['Residential Type'].astype(str).fillna('Unknown')
df['Residential Type'] = df['Residential Type'].replace('nan', 'Unknown')

df['Assessor Remarks'] = df['Assessor Remarks'].astype(str).fillna('Unknown')
df['Assessor Remarks'] = df['Assessor Remarks'].replace('nan', 'Unknown')

df['OPM remarks'] = df['OPM remarks'].astype(str).fillna('Unknown')
df['OPM remarks'] = df['OPM remarks'].replace('nan', 'Unknown')

df['Location'] = df['Location'].astype(str).fillna('Unknown')
df['Location'] = df['Location'].replace('nan', 'Unknown')

df['Non Use Code'] = df['Non Use Code'].astype(str).fillna('Unknown')
df['Non Use Code'] = df['Non Use Code'].replace('nan', 'Unknown')

df['Date Recorded'] = df['Date Recorded'].astype(str).fillna(2000)
df['Date Recorded'] = df['Date Recorded'].replace('nan', 2000)

# check the head of the column
df.head()
```

## Assessment Accuracy Evaluate the accuracy of property assess values with actual sale amounts, and identify any discrepancies
-- Visualise Property Assessment Values with Actual Sales Amount using the Sales Ratio

```python
# Display the first few rows of the dataset to understand its structure
print(df.head())

# Calculate the percentage difference between assessed values and actual sale amounts
df['Percentage_Difference'] = ((df['Assessed Value'] - df['Sale Amount']) / df['Sale Amount']) * 100

# Display the updated DataFrame with the percentage difference column
print(df)

# Set a threshold for identifying discrepancies (e.g., 10% difference)
discrepancy_threshold = 10

# Identify properties with discrepancies beyond the threshold
discrepancies = df[df['Percentage_Difference'].abs() > discrepancy_threshold]

# Display properties with discrepancies
print("Properties with discrepancies:")
print(discrepancies)
```

The Sales Ratio which should be 1 if the Assessed Value and Sale Amount is correct shows a significant discrepany

```python
# Calculate the percentage difference between assessed values and actual sale amounts
df['Percentage_Difference'] = ((df['Assessed Value'] - df['Sale Amount']) / df['Sale Amount']) * 100

# Customized line chart
plt.figure(figsize=(12, 6))
plt.plot(df.index, df['Percentage_Difference'], marker='o', linestyle='-', color='r', label='Percentage Difference')

# Add labels and title
plt.title('Percentage Difference Between Assessed Values and Sale Amounts')
plt.xlabel('Property Index')
plt.ylabel('Percentage Difference')

# Customize the x-axis tick labels (optional)
plt.xticks(rotation=45, ha='right')

# Add a horizontal line at the threshold for better visualization
plt.axhline(y=discrepancy_threshold, color='black', linestyle='--', label='Threshold')
plt.axhline(y=-discrepancy_threshold, color='black', linestyle='--')

# Add legend
plt.legend()

# Show the plot
plt.show()
```

The above chart shows the percentage difference between the Assessed Value and Actual Sale Amount of each Property. The Discrepany went as high as above 1.2%

## Geographical Analysis Investigate variations in assessed values, sales amounts, and market ratios across different towns, providing a localized understanding of real estate dynamics

```python
Columns_to_Analyse = ['Assessed Value', 'Sale Amount', 'Sales Ratio', 'Town']

# Customized bar chart
plt.figure(figsize=(16, 8))
sns.barplot(x='Town', y='Assessed Value', data=df, color='skyblue', label='Assessed Value')
sns.barplot(x='Town', y='Sale Amount', data=df, color='orange', label='Sale Amount')
sns.barplot(x='Town', y='Sales Ratio', data=df, color='green', label='Sales Ratio')

# Add labels and title
plt.title('Comparison of Assessed Value, Sale Amount, and Sales Ratio by Town')
plt.xlabel('Town')
plt.ylabel('Values')

# Show the legend
plt.legend()

# Rotate x-axis labels for better visibility
plt.xticks(rotation=45, ha='right')

# Show the plot
plt.show()
```

```python
# Sort the DataFrame by each column
top_20_assessed_value = df.sort_values(by='Assessed Value', ascending=False).head(20)
top_20_sale_amount = df.sort_values(by='Sale Amount', ascending=False).head(20)
top_20_sales_ratio = df.sort_values(by='Sales Ratio', ascending=False).head(20)

# Create a figure and axes
fig, axes = plt.subplots(3, 1, figsize=(12, 18))

# Customized bar chart function
def customized_bar_chart(data, x_column, y_column, color, title, xlabel, ylabel, ax):
    sns.barplot(x=x_column, y=y_column, data=data, palette=color, ax=ax)
    ax.set_title(title)
    ax.set_xlabel(xlabel)
    ax.set_ylabel(ylabel)

# Customize and plot Assessed Value
customized_bar_chart(top_20_assessed_value, 'Assessed Value', 'Town', 'Blues_d',
                      'Top 20 Towns - Assessed Value', 'Assessed Value', 'Town', ax=axes[0])

# Customize and plot Sale Amount
customized_bar_chart(top_20_sale_amount, 'Sale Amount', 'Town', 'Oranges_d',
                      'Top 20 Towns - Sale Amount', 'Sale Amount', 'Town', ax=axes[1])

# Customize and plot Sales Ratio
customized_bar_chart(top_20_sales_ratio, 'Sales Ratio', 'Town', 'Greens_d',
                      'Top 20 Towns - Sales Ratio', 'Sales Ratio', 'Town', ax=axes[2])

# Adjust layout
plt.tight_layout()

# Show the plot
plt.show()
```

Using the Top 20 Towns to investigate the variations in Assessed Value, Sale Amount and Market Ratio, The above shows that Hamden is the highest in terms of Assessed Value, Willington is the highest in terms of Sale Amount and Salisbury is the highest in terms of Sales Ratio. This clearly shows that there is a great discrepancy between the Assessed Value and Actual Sale Amount.

## Non-Use Code Insights Explore the significance of non-use codes in property assessment, investigating how these codes influence assessed values and sales transactions.

```python
columnsVisualization = ['Non Use Code', 'Assessed Value', 'Sale Amount']

# Create a grouped bar chart
plt.figure(figsize=(12, 8))
sns.barplot(x='Non Use Code', y='Assessed Value', data=df, ci=None, color='blue', label='Assessed Value')
sns.barplot(x='Non Use Code', y='Sale Amount', data=df, ci=None, color='orange', label='Sale Amount')

plt.title('Significance of Non Use Code on Assessed Value and Sale Amount')
plt.xlabel('Non Use Code')
plt.ylabel('Values')
plt.legend()

plt.show()
```

```python
# Sort the DataFrame by 'Assessed Value' and select the top 20 Non Use Codes
top_20_non_use_codes = df.sort_values(by='Assessed Value', ascending=False).head(20)

# Set a larger figure size
plt.figure(figsize=(16, 10))

# Create a grouped bar chart with data labels
sns.barplot(x='Non Use Code', y='Assessed Value', data=top_20_non_use_codes, color='blue', label='Assessed Value')

# Bar for Sale Amount
sns.barplot(x='Non Use Code', y='Sale Amount', data=top_20_non_use_codes, color='orange', label='Sale Amount')

# Add data labels for Assessed Value
for index, value in enumerate(top_20_non_use_codes['Assessed Value']):
    plt.text(index, value, f'{value:,.2f}', va='bottom', ha='center', fontsize=10)

# Add data labels for Sale Amount
for index, value in enumerate(top_20_non_use_codes['Sale Amount']):
    plt.text(index, value, f'{value:,.2f}', va='bottom', ha='center', fontsize=10)

plt.title('Top 20 Non Use Codes - Assessed Value and Sale Amount')
plt.xlabel('Non Use Code')
plt.ylabel('Values')
plt.legend()

plt.show()
```
